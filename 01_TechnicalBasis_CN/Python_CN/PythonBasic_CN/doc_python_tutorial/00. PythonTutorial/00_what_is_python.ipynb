
Python是一种易于学习且功能强大的编程语言。它具有高效的高级数据结构和简单但有效的面向对象编程方法。Python优雅的语法和动态类型，加上其解释性质，使其成为许多领域中在大多数平台上进行脚本编写和快速应用开发的理想语言。

Python解释器和广泛的标准库可从Python[官方网站](https://www.python.org/) 免费获取源代码或二进制形式，适用于所有主要平台，并且可以自由分发。该网站还包含许多免费的第三方Python模块、程序和工具的分发版和指针，以及额外的文档。

Python解释器可以轻松地通过在C或C++（或其他可以从C调用的语言）中实现的新功能和数据类型进行扩展。Python也适合作为可定制应用程序的扩展语言。

Python的标准库提供了广泛的模块和工具，这些模块和工具可以帮助开发者进行文件操作、系统管理、网络通信、数据库交互等。这意味着使用Python，开发者可以快速地进行原型设计和开发，而无需从头开始构建许多基础设施。

Python社区活跃且支持全面，无论是通过在线资源如教程和文档，还是通过社区论坛和聚会，初学者和经验丰富的开发者都能找到支持和资源。此外，由于Python的流行，许多现代技术栈和框架都提供了对Python的支持，包括数据科学、机器学习、网络开发和自动化脚本等领域。

本教程以非正式的方式向读者介绍Python语言和系统的基本概念和特性。手头有一个Python解释器用于实践体验是有帮助的，但所有示例都是自包含的，因此本教程也可以离线阅读。

关于标准对象和模块的描述，请参见Python标准库。Python语言参考提供了语言的更正式定义。要用C或C++编写扩展，请阅读扩展和嵌入Python解释器以及Python/C API参考手册。还有几本书深入介绍了Python。

本教程不试图全面覆盖每一个特性，或者甚至每一个常用特性。相反，它介绍了Python的许多值得注意的特性，并将给你一个关于该语言风格和风味的好主意。阅读之后，你将能够阅读和编写Python模块和程序，并准备好了解更多关于在Python标准库中描述的各种Python库模块。



如果你经常使用电脑工作，最终你会发现有些任务你希望能自动化。例如，你可能想要对大量文本文件执行查找和替换操作，或以复杂的方式重命名和重新排列一系列照片文件。也许你想编写一个小型的自定义数据库、一个专门的图形用户界面应用程序，或一个简单的游戏。

如果你是一名专业软件开发人员，你可能需要使用几个C/C++/Java库，但发现常规的编写/编译/测试/重新编译周期太慢了。也许你正在为这样一个库编写测试套件，发现编写测试代码是一项乏味的任务。或者你已经写了一个程序，希望能使用一种扩展语言，你不想为你的应用设计和实现一个全新的语言。

Python正是你需要的语言。

对于这些任务，你可以编写Unix shell脚本或Windows批处理文件，但shell脚本最擅长的是移动文件和更改文本数据，不适合用于图形用户界面应用或游戏。你可以编写C/C++/Java程序，但即使是初稿程序，也需要大量的开发时间。Python使用起来更简单，可用于Windows、macOS和Unix操作系统，并将帮助你更快地完成工作。

Python简单易用，但它是一种真正的编程语言，为大型程序提供了比shell脚本或批处理文件更多的结构和支持。另一方面，Python还提供了比C更多的错误检查，并且作为一种非常高级的语言，它内置了高级数据类型，如灵活的数组和字典。由于其更通用的数据类型，Python适用于比Awk或甚至Perl更广泛的问题域，但许多事情在Python中至少和在那些语言中一样容易。

Python允许你将程序分解为可以在其他Python程序中重用的模块。它附带了一个大型的标准模块集合，你可以将其作为程序的基础——或作为开始学习Python编程的示例。其中一些模块提供了诸如文件I/O、系统调用、套接字，甚至是图形用户界面工具包（如Tk）的接口。

Python是一种解释型语言，在程序开发期间可以为你节省大量时间，因为不需要编译和链接。解释器可以交互式使用，这使得轻松实验语言的特性、编写一次性程序或在自底向上的程序开发过程中测试函数变得简单。它也是一个方便的桌面计算器。

Python使得程序可以紧凑且易于阅读。用Python编写的程序通常比相当的C、C++或Java程序短得多，原因有几个：

高级数据类型让你可以在单个语句中表达复杂的操作；
语句分组是通过缩进而不是开始和结束括号来完成的；
不需要变量或参数声明。
Python是可扩展的：如果你知道如何用C编程，为解释器添加一个新的内置函数或模块很容易，无论是为了以最高速度执行关键操作，还是将Python程序链接到可能只以二进制形式提供的库（如特定供应商的图形库）。一旦你真正入迷，你可以将Python解释器链接到用C编写的应用程序中，并将其用作那个应用程序的扩展或命令语言。

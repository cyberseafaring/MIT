Gymnasium库中的“[经典控制](https://gymnasium.farama.org/main/environments/classic_control/)”环境是一系列设计用来测试和开发强化学习算法的基础环境，它们模拟了控制理论中的一些经典问题。这些环境因其简单而直观的动态系统模型而广受欢迎，适合于入门学习强化学习以及验证新算法的有效性。以下是一些经典控制环境的详细描述：

### 1. CartPole

- **问题描述**：一个杆子竖直安装在一个可沿直线移动的小车上。目标是通过左右移动小车来保持杆子竖直不倒。
- **观察空间**：包括小车的位置和速度，杆子的角度和角速度。
- **动作空间**：二维离散空间，表示小车的移动方向（左或右）。
- **奖励**：每个时间步，如果杆子保持竖直，就会获得奖励。

### 2. MountainCar

- **问题描述**：一辆小车位于一个U形谷底，目标是驾驶小车到达谷的右侧顶部。小车的引擎不足以一次性爬上山坡，需要来回摆动以获得足够的动量。
- **观察空间**：包括小车的位置和速度。
- **动作空间**：三维离散空间，表示小车的加速方向（向前、静止、向后）。
- **奖励**：达到目标位置时给予奖励，否则每个时间步惩罚。

### 3. Pendulum

- **问题描述**：一个单摆的控制问题，目标是通过施加力矩使摆杆竖直向上。
- **观察空间**：连续空间，包括摆杆的角度和角速度。
- **动作空间**：连续空间，表示施加在摆杆上的力矩。
- **奖励**：根据摆杆的角度和角速度以及施加的力矩计算。

### 4. Acrobot

- **问题描述**：一个由两个杆子组成的机器人臂（简称为双摆），下面的杆子固定在某点，上面的杆子可以自由摆动。目标是通过施加力矩使上面的杆子达到一定高度。
- **观察空间**：包括两个杆子的角度和角速度。
- **动作空间**：三维离散空间，表示施加在连接点上的力矩方向。
- **奖励**：根据上面杆子的高度给予奖励。

这些环境提供了不同的挑战，从简单的线性控制（如CartPole）到更复杂的非线性动态系统（如Acrobot）。它们的共同点在于，都需要学习策略来优化长期奖励，这是强化学习研究的核心目标之一。经典控制环境因其简单性和易于理解的物理直觉，成为了强化学习社区中非常重要的一部分。
